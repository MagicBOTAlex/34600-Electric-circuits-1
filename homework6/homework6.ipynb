# Task 1
Formula: $Q=\frac{1}{2} \cdot C \cdot v^2$ \
We know C and v, so just insert and find Q

```python
from sympy import *
v = -23 # V
c = 3 # uF
q = (1/2)*c*v**2
print(f"Charged {N(q, 3)}uJ")
```

Therefore the charged amount at t=0 is $Q=\boxed{794uJ}$

# Task 2
Formula $\Delta v=\frac{1}{C}\int_{0}^{t_1} i(t)\,dt$ and $v_C(t_1)=v_C(0)+\Delta v$, but $i(t)$ is linear, therfore $\frac{1}{C} \cdot \frac{i_0+i_1}{2}\cdot \Delta t = \Delta v$

We know: \
$v_c(0)=-23.0V$ \
$i_0=-23.0A$ \
$t_1=10.0uS$ \
Now insert and find $v_c(t_1)$

```python
from sympy import *
i_0 = -23 # A
i_1 = -5 # A
t_1 = 10e-6 # S
c = 3e-6 # F
v0 = -23 # V
v1 = (1/c) * ((i_0+i_1)/2) * t_1
v_t1 = v0 + v1
print(f"V(t1): {N(v_t1, 3)}V")
```

Therefore the capasitor voltage across at $t_1$ is $\boxed{-69.7V}$

# Task 3

```python
from sympy import *
i_0 = -5 # A
i_1 = -23 # A
t_0 = 10e-6 # S
t_1 = 15e-6 # S
c = 3e-6 # F
v2 = (1/c) * ((i_0+i_1)/2) * (t_1-t_0)
v_t2 = v_t1 + v2
print(f"V(t2): {N(v_t2, 3)}V")

from sympy import *
i_0 = -23 # A
i_1 = -23 # A
t_0 = 15e-6 # S
t_1 = 20e-6 # S
c = 3e-6 # F
v3 = (1/c) * ((i_0+i_1)/2) * (t_1-t_0)
v_t3 = v_t2 + v3
print(f"V(t2): {N(v_t3, 3)}V")
```

Therefore: $v_C(t_3)=\boxed{-131V}$

# Task 4

# Task 5
We have these key points: \
$v_C(0)= -23.0\text{ V}$ \
$v_C(t_1=10\,\mu s)= -69.7\text{ V}$ \
$v_C(t_2=15\,\mu s)= -93.0\text{ V}$ \
$v_C(t_3=20\,\mu s)= -131\text{ V}$ \

```python
import matplotlib.pyplot as plt
import numpy as np

t_us = np.array([0, 10, 15, 20], dtype=float)  # uS
vC   = np.array([-23.0, -69.7, -93.0, -131]) # V

plt.figure()
plt.plot(t_us, vC, marker='o', linewidth=2)
plt.xlabel('Time (us)')
plt.ylabel('v_C (V)')
plt.title('Capacitor Voltage vs Time')
plt.grid(True, linestyle='--', alpha=0.5)

for x, y in zip(t_us, vC):
    plt.annotate(f'{y:.1f} V', (x, y), textcoords='offset points', xytext=(6, 6))

plt.tight_layout()
plt.show()
```

### Answer
$[0,t_1]$: Current increases linearly, while voltage would steadily decrease in a parabola shape towards -69.7V\
$[t_1,t_2]$: Current decreases linearly, while voltage would steadily increase in a parabola shape towards -93.0V\
$[t_2,t_3]$: Current stays constant, while voltage would steadily decrease linearly towards -131V\

# Task 10


```python
import matplotlib.pyplot as plt
import numpy as np

t_us = np.array([0.0, 10.0, 15.0, 20.0])

vL_levels = np.array([5.4, -10.8, 0.0])

vL_for_step = np.r_[vL_levels, vL_levels[-1]]

plt.figure()
plt.step(t_us, vL_for_step, where='post', linewidth=2)
plt.scatter(t_us, vL_for_step, zorder=3)

plt.xlabel('Time (us)')
plt.ylabel('v_L (V)')
plt.title('Inductor Voltage vs Time')
plt.grid(True, linestyle='--', alpha=0.5)

midpoints = (t_us[:-1] + t_us[1:]) / 2
for x, y in zip(midpoints, vL_levels):
    plt.annotate(f'{y:.1f} V', (x, y), textcoords='offset points', xytext=(6, 6))

plt.tight_layout()
plt.show()
```

### Answer
$[0, t_1]$: (v = +5.40) V (flat) \
$[t_1, t_2]$: (v = -10.8) V (flat) \
$[t_2, t_3]$: (v = 0) V (flat)
