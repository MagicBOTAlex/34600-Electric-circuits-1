```python
from sympy import *

R_1 = 289
R_4 = 121 
v_4 = 5.6
```

# task 1
 \
We know voltage and resistance of R_4, therefore \
$V_4=I_4\cdot R_4$

```python
x = symbols("x")
I_4 = N(solve(Eq(v_4, x*R_4), x)[0], 3)
print(f"I_4 is {I_4}A")
```

# Task 2
 \
Formula: $P=I^2  \cdot R$


```python
P_4 = N(I_4**2*R_4, 3)
print(f"P_4 is {P_4}W")
```

# Task 3
 \
Since $I_4 = I_3$ because of kirchoff \
Then using: $P = I^2 \cdot R$ \
And the task says: $P_4 \cdot 3 = P_3$, therefore:
$I_4^2 \cdot R_4 \cdot 3 = I_3^2 \cdot R_3$ \
Insert and find $R_3$

```python
I_3 = I_4
eq = Eq(I_4**2 * R_4 * 3, I_4**2 * x)
R_3 = N(solve(eq, x)[0], 3)
print(f"Therefore, R_3 is {R_3} Ohms")
```

# Task 4
 \
Find $v_x$ \
 \
$v_3 = I_3^2\cdot R_3$ \
$v_x = v_4 + v_3$ \
Task says $P_2=2\cdot (P_3+P_4)$

```python
P_3 = I_3**2 * R_3
P_2 = N(2*(P_3 + P_4), 3)
v_3 = N(I_3 * R_3, 3)
v_x = N(v_4 + v_3, 3)
print(f"R_2 power dissipation: {P_2}W")
print(f"v_x: {v_x}v")
```

$I_2=\frac{P_2}{v_x}$ \
$R_2=\frac{v_x}{I_2}$

```python
I_2 = N(P_2 / v_x, 3)
R_2 = N(v_x / I_2, 3)
print(f"Therefore R_2={R_2} Ohms")
```

# Task 5
We need $I_1$ first, and this is done with: \
$I_1 = \frac{v_x}{R_1}$

```python
I_1 = N(v_x / R_1, 3)
print(f"I_1 current is: {I_1}A")
```

Now using kirchoff something rule: \
$I_0 = I_1 + I_2 + I_3$ \
We're using $I_3$ because $R_3$ and $R_4$ are in serial, and the current doesn't change.

```python
I_0 = I_1 + I_2 + I_3
print(f"I_0={I_0}A")
```
